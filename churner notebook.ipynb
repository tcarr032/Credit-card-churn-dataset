# Heat map exploratory data # 

First we are going to check specifically for positive correlations with in the data

```python
import matplotlib.pyplot as plt
import seaborn as sb

plt.figure(figsize=(20,15))
dataplot = sb.heatmap(corr,fmt='.2f', cmap='GnBu', annot=True, vmin=0)
plt.show()
```

There are correlations between the following:

Months_on_book : Customer_Age

Avg_Utilization_Ration : Total_Revolving_Bal

Total_Trans_Amt : Total_Trans_ct

Total_Ct_Chng_Q4_Q1 : Total_Amt_Chng_Q4_Q1

We will graph these out in a scatterplot and see where the attritioned customers lie on the graphs


```python
fig, ax = plt.subplots(2,2, figsize=(20,30))
sb.scatterplot(df['Months_on_book'], df['Customer_Age'], hue=df['Attrition_Flag'], ax=ax[0,0])
sb.scatterplot(df['Avg_Utilization_Ratio'], df['Total_Revolving_Bal'], hue=df['Attrition_Flag'], ax=ax[0,1])
sb.scatterplot(df['Total_Trans_Amt'], df['Total_Trans_Ct'], hue=df['Attrition_Flag'], ax=ax[1,0])
sb.scatterplot(df['Total_Ct_Chng_Q4_Q1'], df['Total_Amt_Chng_Q4_Q1'], hue=df['Attrition_Flag'], ax=ax[1,1])
plt.show()
```

from the plots there is a positivie correlation with all plot with positive relation with attrition customers with everyting except customer age and Months on book

customer age and Months on book does not show any sort of insight of atrophied customers

```python
import pandas as pd

df = pd.read_csv("BankChurners.csv")
```

```python
df.head()
```

```python
#replace long Naive Bayes columns
df.rename(columns={df.columns[-2]: "NBayes_attrition1", df.columns[-1]: "NBayes_attrition2"}, inplace=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
for col in df.columns:
    if df.loc[:, col].dtype == 'object':
        print(f'{col}: {df.loc[:,col].unique()} = {len(df.loc[:,col].unique())}')
```

```python
#make categories from objects
for col in df.columns:
    if df.loc[:, col].dtype == 'object':
        df.loc[:, f'{col}_code'] = df.loc[:, col].astype('category').cat.codes
```

```python
df.head()
```

```python
corr = df.corr()
corr
```

# Heat map exploratory data # 

First we are going to check specifically for positive correlations with in the data

```python
import matplotlib.pyplot as plt
import seaborn as sb

plt.figure(figsize=(20,15))
dataplot = sb.heatmap(corr,fmt='.2f', cmap='GnBu', annot=True, vmin=0)
plt.show()
```
